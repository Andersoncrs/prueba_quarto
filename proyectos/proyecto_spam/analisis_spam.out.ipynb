# Análisis Experimental de Detección de Spam

Este cuaderno contiene el análisis técnico y entrenamiento del modelo.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Simulación de una matriz de confusión para Spam
data = np.array([[850, 15], [30, 115]])
sns.set_theme(style='white')
plt.figure(figsize=(5, 4))
sns.heatmap(data, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.title('Matriz de Confusión del Clasificador')
plt.ylabel('Realidad')
plt.xlabel('Predicción')
plt.show()
```
