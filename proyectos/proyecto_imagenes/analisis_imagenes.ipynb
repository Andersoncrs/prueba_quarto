# Entrenamiento de Clasificador de Imágenes (CNN)
Prueba experimental de curvas de pérdida y métricas de validación.

```python
#| label: fig-loss-curves
#| fig-cap: 'Curva de pérdida (Loss) durante las épocas de entrenamiento'
#| warning: false

import matplotlib.pyplot as plt

# Simulación de curvas de entrenamiento
epochs = list(range(1, 11))
train_loss = [0.8, 0.6, 0.45, 0.32, 0.25, 0.19, 0.15, 0.12, 0.10, 0.08]
val_loss = [0.85, 0.65, 0.52, 0.41, 0.35, 0.31, 0.29, 0.28, 0.27, 0.27]

plt.figure(figsize=(6, 4))
plt.plot(epochs, train_loss, label='Entrenamiento', color='blue', marker='o')
plt.plot(epochs, val_loss, label='Validación', color='red', marker='x')
plt.title('Historial de Pérdida (Loss Curve)')
plt.xlabel('Época')
plt.ylabel('Pérdida (Cross-Entropy)')
plt.legend()
plt.grid(True)
plt.show()
```
